```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import calendar
```

# Dataset link:
## https://www.kaggle.com/datasets/itssuru/global-terrorism

```python
# Loading the Dataset
df=pd.read_csv("Global_Terrorism.csv",encoding='ISO-8859-1')
df
```

```python
# Display the shape of the Dataset
print("There are {} rows and {} columns in the Dataset".format(df.shape[0],df.shape[1]))
```

```python
#Display the columns
df.columns.values
```

```python
#Renaming the names of columns which we will be requiring for our Dataset
df=df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                             'provstate':'State','city':'City','success':'Success','suicide':'Suicide','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type',
                             'targsubtype1_txt':'Target_subtype','gname':'Gang_name','weaptype1_txt':'Weapon_type'})

df
```

```python
df.describe().T
```

```python
#Taking only the columns which we will require for our Dataset
df=df[['Year','Month','Day','Country','Region','State','City','Success','Suicide','Attack_type','Target_type','Target_subtype','Gang_name','Weapon_type']]
df
```

```python
#Checking for the null values in the Dataset
df.isnull().sum()
```

```python
#Replacing the null values with Unknown and Others
df['State']=df['State'].fillna('Unknown')
df['City']=df['City'].fillna('Unknown')
df['Target_subtype']=df['Target_subtype'].fillna('Others')
```

```python
#Displaying the table after removing all the null values
df.isnull().sum()
```

```python
print(f"Maximum Attacks in Year      : {df['Year'].value_counts().idxmax()}")
print(f"Maximum Attacks in Month     : {calendar.month_name[df['Month'].value_counts().idxmax()]}")
print(f"Maximum Attacks in Country   : {df['Country'].value_counts().idxmax()}")
print(f"Maximum Attacks in Region    : {df['Region'].value_counts().idxmax()}")
print(f"Maximum Attacks in City      : {df['City'].value_counts().keys()[1]}")
print(f"Success Rate of Attacks      : {(df['Success'].value_counts()[1])*100//len(df['Success'])} %")
print(f"Rate of Non-Suicidal Attacks : {(df['Suicide'].value_counts()[0])*100//len(df['Suicide'])} %")
print(f"Most Attack Type             : {df['Attack_type'].value_counts().idxmax()}")
print(f"Most Target Type             : {df['Target_type'].value_counts().idxmax()}")
print(f"Most Target Sub-Type         : {df['Target_subtype'].value_counts().idxmax()}")
print(f"Maximum Attacking Group      : {df['Gang_name'].value_counts().keys()[1]}")
print(f"Maximum Attacking Weapon     : {df['Weapon_type'].value_counts().idxmax()}")
```

```python
#Total number of countries which were recorded in the Dataset
Total_Countries=df['Country'].nunique()
print("There were %d countries which were affected"%Total_Countries)
```

# Visualization

```python
x_axis = df['Year'].value_counts().keys()
y_axis = df['Year'].value_counts().values

plt.figure(figsize=(16,7))
plt.title('Terrorist Attacks in various Years', fontsize=20)
plt.xticks(df['Year'].unique(), rotation=45)
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)

sns.barplot(x=x_axis,y=y_axis)
```

```python
x_axis = df['Country'].value_counts().keys()[:20]
y_axis = df['Country'].value_counts().values[:20]

plt.figure(figsize=(16,7))
plt.title('Terrorist Attacks in various Countries', fontsize=20)
plt.xticks(df['Year'].unique(), rotation=45)
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)

sns.barplot(x=x_axis,y=y_axis)
```

```python
x_axis = df['State'].value_counts().keys()[:20]
y_axis = df['State'].value_counts().values[:20]

plt.figure(figsize=(16,7))
plt.title('Terrorist Attacks in various States', fontsize=20)
plt.xticks(df['Year'].unique(), rotation=45)
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)

sns.barplot(x=x_axis,y=y_axis)
```

```python
x_axis = df['City'].value_counts().keys()[1:20]
y_axis = df['City'].value_counts().values[1:20]

plt.figure(figsize=(16,7))
plt.title('Terrorist Attacks in various Cities', fontsize=20)
plt.xticks(df['Year'].unique(), rotation=45)
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)

sns.barplot(x=x_axis,y=y_axis)
```

```python
plt.figure(figsize = (16,9))
ax = sns.countplot(df['Year'], palette = 'rocket_r')
plt.xticks(rotation = 90)
plt.ylabel("No of Attacks") 
plt.title("Year-Wise Attacks")

for p in ax.patches:
    ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.15, p.get_height()+100), rotation = 90, size =12)

plt.show()
```

```python
x_axis = df['Target_type'].value_counts().keys()
y_axis = df['Target_type'].value_counts().values

plt.figure(figsize=(16,7))

plt.title('Different Target Types', fontsize=20)
sns.barplot(x=x_axis,y=y_axis)
plt.xticks(rotation=60)
plt.xlabel('Target',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)
```

```python
x_axis = df['Gang_name'].value_counts().keys()[1:21]
y_axis = df['Gang_name'].value_counts().values[1:21]

plt.figure(figsize=(16,7))

plt.title('Top 20 Terrorist groups', fontsize=20)
sns.barplot(x=x_axis,y=y_axis)
plt.xticks(rotation=90)
plt.xlabel('Group Name',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)
```

```python
x_axis = df['Weapon_type'].value_counts().keys()[1:10]
y_axis = df['Weapon_type'].value_counts().values[1:10]

plt.figure(figsize=(16,7))

plt.title('Different TYpes of Weapons', fontsize=20)
sns.barplot(x=x_axis,y=y_axis)
plt.xticks(rotation=90)
plt.xlabel('Weapon',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)
```
